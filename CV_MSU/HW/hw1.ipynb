```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
image = cv2.imread("Apollo_11_Launch.jpg", 1)
image_copy = image.copy()
for q in range(10):
    cv2.circle(image_copy, (600, 400),100 + q *20, (0, 100, 100), thickness = 10)

plt.imshow(image_copy[:,:,::-1])
```

```python
image_copy = image.copy()

for q in range(10):
    cv2.rectangle(image_copy, (500 + 10 * q, 100 + 10 * q), (700 - 10 * q, 700 - 10 * q), (255, 100, 100), thickness = 3)
plt.imshow(image_copy[:,:,::-1])
```

```python
image_copy = image.copy()

cv2.circle(image_copy, (600, 400),300, (0, 100, 100), thickness = 10)
cv2.rectangle(image_copy, (400, 180), (800, 620), (255, 100, 100), thickness = 10)

cv2.line(image_copy, (600, 180), (800, 500), (0, 0, 255), thickness = 10)
cv2.line(image_copy, (800, 500), (500, 620), (0, 0, 255), thickness = 10)
cv2.line(image_copy, (500, 620), (600, 180), (0, 0, 255), thickness = 10)

plt.imshow(image_copy[:,:,::-1])
```

```python
image_copy = image.copy()

cv2.rectangle(image_copy, (400, 180), (800, 620), (255, 100, 100), thickness = 10)
cv2.circle(image_copy, (600, 400),200, (0, 100, 100), thickness = 10)

cv2.line(image_copy, (600, 200), (770, 500), (0, 0, 255), thickness = 10)
cv2.line(image_copy, (770, 500), (500, 570), (0, 0, 255), thickness = 10)
cv2.line(image_copy, (500, 570), (600, 200), (0, 0, 255), thickness = 10)

plt.imshow(image_copy[:,:,::-1])
```

```python
image_copy = image.copy()
x = image_copy.shape[0]
y = image_copy.shape[1]
x2 = int(x/2)
y2 = int(y/2)

matBright = image_copy.copy()
matBright = np.int32(matBright)
matBright[:, 0:y2] -= 30
matBright[:, y2:y] += 30
matBright = np.uint8(np.clip(matBright,0,255))
plt.imshow(matBright[:,:,::-1])

```

```python
image_copy = image.copy()
x = image_copy.shape[0]
y = image_copy.shape[1]
x2 = int(x/2)
y2 = int(y/2)

mat1 = np.ones(image_copy.shape) * 0.7
mat2 = np.ones(image_copy.shape) * 1.3
imgRgbLow = np.uint8(cv2.multiply(np.float64(image_copy), mat1)) 
imgRgbHigh = np.uint8(np.clip(cv2.multiply(np.float64(image_copy), mat2), 0, 255))

image_copy[0:x2, 0:y] = imgRgbHigh[0:x2, 0:y]
image_copy[x2:x, 0:y] = imgRgbLow[x2:x, 0:y]
plt.imshow(image_copy[:,:,::-1])
```

```python
image_copy = cv2.imread("picture/New_Zealand_Lake.jpg", 0)
retval, binBuild = cv2.threshold(image_copy, 180, 255, cv2.THRESH_TRUNC)
binBuildAdaptive = cv2.adaptiveThreshold(image_copy, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 17, -2)
plt.figure(figsize=[18, 5])
plt.subplot(121)
plt.imshow(binBuild, cmap='gray')

plt.subplot(122)
plt.imshow(binBuildAdaptive, cmap='gray')
```
